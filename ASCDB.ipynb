```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('ASCDB_UE_C.csv')
UE = data[['B3LYP', 'B3PW91', 'B97-1', 'B97-2', 'B97-D3(0)', 'BLYP', 'BMK', 'BP86', 'HCTH/407', 'HSE-HJS', 'M05', 'M05-2X', 'M06', 'M06-2X', 'M06-L', 'M08-HX', 'M11', 'M11-L', 'MN12-L', 'MN12-SX', 'MN15', 'MN15-L', 'N12', 'N12-SX', 'OLYP', 'PBE', 'PBE0', 'PW6B95', 'PW91', 'revPBE', 'revTPSS', 'RPBE', 'SOGGA11-X', 'τ-HCTH', 'τ-HCTHh', 'TPSS', 'TPSSh', 'B97M-V', 'B97M-rV', 'ωB97X-V', 'ωB97M-V', 'B3LYP-D3(BJ)', 'PBE-D3(BJ)', 'PBE0-D3(BJ)', 'PW6B95-D3(BJ)', 'SCAN-D3(BJ)', 'SVWN5', 'HF', 'revPBE-D3(BJ)', 'M06-L-D3(0)', 'B2PLYP-D3(BJ)', 'DSD-PBEP86-D3(BJ)', 'PWPB95-D3(BJ)','revTPSS-D3(BJ)','M11-D3(BJ)','M11-L-D3(0)','M06-D3(0)','M06-2X-D3(0)','N12-D3(0)','N12-SX-D3(BJ)']]
```

```python
# Number of data-point in each subset/class:
NCA15_N  = 15
NCB21_N  = 21
NCC21_N  = 21
NCW3_N   = 3
TARE19_N = 19
TBHA20_N = 20
TBHB16_N = 16
THR5_N   = 5
NLMA15_N = 15
NLMB7_N  = 7
NLMR11_N = 11
NLSIE9_N = 9
UMBA8_N  = 8
UMBB5_N  = 5
UMBC7_N  = 7
UAE18_N  = 18

NC_N = NCA15_N + NCB21_N + NCC21_N + NCW3_N
TC_N = TARE19_N + TBHA20_N + TBHB16_N + THR5_N
NL_N = NLMA15_N + NLMB7_N + NLMR11_N + NLSIE9_N
U_N = UMBA8_N + UMBB5_N + UMBC7_N + UAE18_N

# Set the location of each subset/class in the CSV:
NCA15_LOC  = 1
NCB21_LOC  = NCA15_LOC + NCA15_N 
NCC21_LOC  = NCB21_LOC + NCB21_N 
NCW3_LOC   = NCC21_LOC + NCC21_N 
TARE19_LOC = NCW3_LOC + NCW3_N  
TBHA20_LOC = TARE19_LOC + TARE19_N
TBHB16_LOC = TBHA20_LOC + TBHA20_N
THR5_LOC   = TBHB16_LOC + TBHB16_N
NLMA15_LOC = THR5_LOC + THR5_N  
NLMB7_LOC  = NLMA15_LOC + NLMA15_N
NLMR11_LOC = NLMB7_LOC + NLMB7_N 
NLSIE9_LOC = NLMR11_LOC + NLMR11_N
UMBA8_LOC  = NLSIE9_LOC + NLSIE9_N
UMBB5_LOC  = UMBA8_LOC + UMBA8_N 
UMBC7_LOC  = UMBB5_LOC + UMBB5_N 
UAE18_LOC  = UMBC7_LOC + UMBC7_N 

NC_LOC = 1
TC_LOC = NC_LOC + NC_N
NL_LOC = TC_LOC + TC_N
UC_LOC = NL_LOC + NL_N

dloc    = UE.loc[0,:]

NCA15    = UE.loc[NCA15_LOC:NCB21_LOC-1,:]
NCB21    = UE.loc[NCB21_LOC:NCC21_LOC-1,:]
NCC21    = UE.loc[NCC21_LOC:NCW3_LOC-1,:]
NCW3     = UE.loc[NCW3_LOC:TARE19_LOC-1,:]
TARE19   = UE.loc[TARE19_LOC:TBHA20_LOC-1,:]
TBHA20   = UE.loc[TBHA20_LOC:TBHB16_LOC-1,:]
TBHB16   = UE.loc[TBHB16_LOC:THR5_LOC-1,:]
THR5     = UE.loc[THR5_LOC:NLMA15_LOC-1,:]
NLMA15   = UE.loc[NLMA15_LOC:NLMB7_LOC-1,:]
NLMB7    = UE.loc[NLMB7_LOC:NLMR11_LOC-1,:]
NLMR11   = UE.loc[NLMR11_LOC:NLSIE9_LOC-1,:]
NLSIE9   = UE.loc[NLSIE9_LOC:UMBA8_LOC-1,:]
UMBA8    = UE.loc[UMBA8_LOC:UMBB5_LOC-1,:]
UMBB5    = UE.loc[UMBB5_LOC:UMBC7_LOC-1,:]
UMBC7    = UE.loc[UMBC7_LOC:UAE18_LOC-1,:]
UAE18    = UE.loc[UAE18_LOC:,:]

NC = UE.loc[NC_LOC:TC_LOC-1,:]
TC = UE.loc[TC_LOC:NL_LOC-1,:]
NL = UE.loc[NL_LOC:UC_LOC-1,:]
UC = UE.loc[UC_LOC:,:]

# Weights for calculating mCV:
NCA15_wi  = 0.396357538
NCB21_wi  = 0.073726349
NCC21_wi  = 0.164612094
NCW3_wi   = 0.034031965
TARE19_wi = 0.05981155
TBHA20_wi = 0.042125533
TBHB16_wi = 0.052209978
THR5_wi   = 0.003169408
NLMA15_wi = 0.004107248
NLMB7_wi  = 0.02005333
NLMR11_wi = 0.047458458
NLSIE9_wi = 0.056366729
UMBA8_wi  = 0.015696887
UMBB5_wi  = 0.019800469
UMBC7_wi  = 0.009620408
UAE18_wi  = 0.000852056

wMUE = NCA15_wi*NCA15.mean() + NCB21_wi*NCB21.mean() + NCC21_wi*NCC21.mean() + NCW3_wi*NCW3.mean() + TARE19_wi*TARE19.mean() + TBHA20_wi*TBHA20.mean() + TBHB16_wi*TBHB16.mean() + THR5_wi*THR5.mean() + NLMA15_wi*NLMA15.mean() + NLMB7_wi*NLMB7.mean() + NLMR11_wi*NLMR11.mean() + NLSIE9_wi*NLSIE9.mean() + UMBA8_wi*UMBA8.mean() + UMBB5_wi*UMBB5.mean() + UMBC7_wi*UMBC7.mean() + UAE18_wi*UAE18.mean()

AIC = wMUE + (2*dloc/200)*(200/(200-dloc))*wMUE
VCC = wMUE * 1/(1-np.sqrt((dloc/200)-(dloc/200)*np.log(dloc/200)+np.log(200)/(2*200)))
CVC = wMUE+dloc/100*UC.mean()/wMUE


AIC_r = AIC.rank() 
VCC_r = VCC.rank() 
CVC_r = CVC.rank() 

AVG_r = (AIC_r+VCC_r+CVC_r)/3


```

```python
# Fill the (SC) table with statistical coefficients
SC = wMUE.to_frame()
SC=SC.rename(columns= {0: 'wMUE'})
SC.insert(0, "DoF", dloc, True)
#SC['NC']  = NC.mean().to_frame()
#SC['TC']  = TC.mean().to_frame()
#SC['NL']  = NL.mean().to_frame()
#SC['UC']  = UC.mean().to_frame()
SC['AIC_t'] = AIC.to_frame()
SC['AIC_r'] = AIC_r.to_frame()

SC["AIC"] = SC["AIC_t"].round(2).map(str) +" ("+SC["AIC_r"].map(int).map(str)+")"
del SC['AIC_t']
del SC['AIC_r']

SC['VCC_t'] = VCC.to_frame()
SC['VCC_r'] = VCC_r.to_frame()

SC["VCC"] = SC["VCC_t"].round(2).map(str) +" ("+SC["VCC_r"].map(int).map(str)+")"
del SC['VCC_t']
del SC['VCC_r']

SC['CVC_t'] = CVC.to_frame()
SC['CVC_r'] = CVC_r.to_frame()

SC["CVC"] = SC["CVC_t"].round(2).map(str) +" ("+SC["CVC_r"].map(int).map(str)+")"
del SC['CVC_t']
del SC['CVC_r']



SC['AVG ranking'] = AVG_r.to_frame()



# drop every functional with wMUE > 8.0
#drop_wMUE = SC[ SC['wMUE'] >= 8.0 ].index
#SC.drop(drop_wMUE , inplace=True)

# drop every functional with sTI > 3.0
#drop_sTI = SC[ SC['sTI'] >= 3.0 ].index
#SC.drop(drop_sTI , inplace=True)

SC=SC.round(2)
SC = SC.sort_values('AVG ranking', ascending=True)
SC

SC.to_csv('Table1.csv')
```
